```python
sentence = input("Enter sentence: ")

words = sentence.split()

unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

# Loop to check each price
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

```python
# Attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count present days
present_count = attendance.count("P")

# Total days
total_days = len(attendance)

# Calculate percentage
percentage = (present_count / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

```python
# List of phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Remove duplicates using set
unique_numbers = set(phone_numbers)

# Print result
print("Unique phone numbers:", unique_numbers)
```

```python
# Input string
text = "python"

# Empty dictionary
char_count = {}

# Loop through each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

# Print result
print(char_count)
```

```python
# Given list
numbers = [10, 20, 30]

# Convert list to tuple
numbers_tuple = tuple(numbers)

# Print result
print("Tuple:", numbers_tuple)
```

```python
# Sample dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Key to check
key_to_check = "Ravi"

# Check if key exists
if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
marks = [70, 75, 80, 65]

# Calculate total marks
total = sum(marks)

# Calculate number of subjects
count = len(marks)

# Calculate average
average = total / count

# Print result
print("Average Marks:", average)
```
